## Sentiment Analysis Dataset Overview and Verification

The dataset contains **community chat messages** that have been preprocessed for sentiment analysis.  

| Column              | Description                                     | Data Type |
|--------------------|-------------------------------------------------|-----------|
| text_message        | Original chat messages from community members  | String    |
| Sentiment           | Sentiment label derived from NLP analysis      | String    |
| Positive_Count      | Count of positive words in the message         | Integer   |
| Negative_Count      | Count of negative words in the message         | Integer   |
| Sentiment_TextBlob  | Sentiment polarity using TextBlob library      | String    |

### Cleaning and Analysis Workflow
1. **NLP Text Preprocessing:**  
   - Removed punctuation, special characters, and numbers.  
   - Tokenized the text and removed stopwords.  
   - Handled negations to capture sentiment correctly.  
   - Counted positive and negative words for rule-based sentiment scoring.  
   - Applied **TextBlob** for additional sentiment polarity verification.  

2. **Verification and Correlation Analysis:**  
   - Converted categorical columns to numerical codes for analysis.  
   - Calculated correlation between numeric sentiment-related columns.  
   - Verified data consistency and ensured no missing or invalid entries remain.  

### Summary
- Raw chat messages were **cleaned and analyzed using NLP techniques**.  
- Python verification ensures the dataset is **accurate, consistent, and ready for further analysis**.  
- Correlation analysis confirms relationships between sentiment counts and derived polarity scores, ensuring **data reliability** for insights.
