## Dataset Overview and Verification

The dataset contains **community chat conversations** related to workouts and diet, gathered from group discussions.  

| Column               | Description                          | Data Type |
|---------------------|--------------------------------------|-----------|
| text_message         | Chat messages from community members | String    |
| cleaned_text         | Tokenized and preprocessed text      | List      |
| workout_related_words| Count of workout-related words       | Integer   |
| diet_related_words   | Count of diet-related words          | Integer   |

### Cleaning Workflow
1. **NLP Cleaning:**  
   - Removed punctuation, special characters, and numbers.  
   - Tokenized text into words for analysis.  
   - Removed stopwords (common words with low meaning).  
   - Handled negations to capture sentiment accurately.  
   - Counted workout- and diet-related keywords for each message.  

2. **Verification in Python (pandas):**  
   - Checked for missing or null values in all columns.  
   - Verified datatype consistency.  
   - Ensured processed columns (`cleaned_text`, keyword counts) match expected format.  
   - Confirmed **no duplicate messages** remain.  

### Summary
- Raw chat data was cleaned using **NLP preprocessing** techniques.  
- Python verification ensures the dataset is **accurate, structured, and ready for analysis**.  
- Final dataset allows **insights on workout vs diet discussions** in the community.
